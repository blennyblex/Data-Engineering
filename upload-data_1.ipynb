```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(taxi_data, name = "yellow_taxi_data", con = engine))
```

```python
df_iter = pd.read_parquet("yellow_tripdata_2025-01.parquet", iterator = True, chunksize = 100000)

df = next(df_iter)
```

```python
df.head(n=0).to_sql(name = "yellow_taxi_data", con = engine, if_exists = "replace")
```

```python
from time import time
```

```python
while True:
    t_start time()
    
    df = next(df_iter)
    
    df.to_sql(name = "yellow_taxi_data", con = engine, if_exists = "append")
    
    t_end = time()
    
    print("inserted another chunck..., took %.3f second"% (tend - t_start))
    
```
